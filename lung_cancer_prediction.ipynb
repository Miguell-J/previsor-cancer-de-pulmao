```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("survey lung cancer.csv")
data
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
plt.figure(figsize=(15, 10))
sns.heatmap(data.corr(),annot=True,cmap='magma')
```

```python
data.iloc[0]
```

```python
colunas = ['GENDER', 'LUNG_CANCER',]
for coluna in colunas:
    data.loc[data[coluna]=='YES', coluna] = 1
    data.loc[data[coluna]=='NO', coluna] = 0
    data.loc[data[coluna]=='M', coluna] = 1
    data.loc[data[coluna]=='F', coluna] = 0
print(data.iloc[0])
display(data)
```

```python
colls = data.columns.tolist()[2:-1]

for col in colls:
    data[col] = data[col].map({1:0, 2:1})
display(data)
```

```python
data.dtypes
```

```python
data['GENDER'] = data['GENDER'].astype(np.int32, copy=False)
data['LUNG_CANCER'] = data['LUNG_CANCER'].astype(np.int32, copy=False)

print(data.dtypes)
```

```python
lista_score = []
x = data.drop('LUNG_CANCER', axis = 1)
y = data['LUNG_CANCER']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)

y_train
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(random_state=42)

lr.fit(x_train, y_train)
```

```python
preds = lr.predict(x_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score, f1_score

lr_score = classification_report(y_test, preds)

print(lr_score)

lista_score.append(accuracy_score(y_test, preds))
```

```python
from sklearn.svm import SVC

svc = SVC()

svc.fit(x_train, y_train)
```

```python
preds2= svc.predict(x_test)

svc_score = classification_report(y_test, preds2)

print(svc_score)

lista_score.append(accuracy_score(y_test, preds2))
```

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()

gnb.fit(x_train, y_train)
```

```python
preds3 = gnb.predict(x_test)

gnb_score = classification_report(y_test, preds3)

print(gnb_score)

lista_score.append(accuracy_score(y_test, preds3))
```

```python
models = ['Logistic Regression', 'Gaussian Naive Bayes', 'SVC']

plt.figure(figsize=(10, 6))
plt.bar(models, lista_score, color='red')
plt.xlabel('Modelos de ML')
plt.ylabel('Acurácia')
plt.title('Comparação de Acurácia ')

plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```
